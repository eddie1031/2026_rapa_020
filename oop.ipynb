## 클래스 메서드, 인스턴스 메서드, 스태틱 메서드(정적 메서드)

- 인스턴스 메서드: 일반적으로 사용하는 메서드, 인스턴스에 귀속되어있는 메서드, 인자로 self를 받는다.
- 클래스 메서드: 인스턴스가 아닌 클래스 자체에 귀속되어있는 메서드. 데코레이터(`@classmethod`)를 추가. 인자로 cls를 받는다.
- 스태틱 메서드: 클래스, 인스턴스 둘 다 귀속되지 않는 메서드. 데코레이터(`@staticmethod`)를 추가. self, cls다 안받는다.

스태틱 메서드는 클래스와 "논리적으로"는 관련이 있습니다. 그러나 클래스나 인스턴스의 상태에 접근이 필요없는 기능에 사용하는 메서드입니다.


```python
class Rectangle:

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

rect = Rectangle(2, 2)
print(rect.area())
```

```python
class User:
    count: int = 0

    def __init__(self, name):
        self.name = name
        User.count += 1

    @classmethod
    def get_count(cls):
        return cls.count

u = User('김민수')

print(User.get_count())
```

```python
# static method

class MathHelper:

    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod
    def is_even(n):
        return n % 2 == 0

print(MathHelper.add(1, 2))
print(MathHelper.is_even(4))
```

# 매직메서드(Magic Method)

메서드 이름 앞뒤에 이중 밑줄(`__`)로 감싸져있는 특별한 메서드를 의미합니다.

직접 호출하는게 아니라, 파이썬의 특정 문법, 내장 함수가 사용될때 자동으로 호출.



```python
class Labtop:

    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price = price

    # toString()
    def __str__(self):
        return f'{self.brand} {self.model} {self.price}원'

labtop = Labtop('삼성', '갤럭시북프로', 100)
print(labtop)
print(str(labtop))
```

```python
class Point:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    # toString(), 사람이 읽기 좋은 표현
    def __str__(self):
        return f'({self.x}, {self.y})'

    # repr -> 개발자를 위한 공식적인 문자열을 표현하는 메서드
    def __repr__(self):
        return f'Point(x={self.x}, y={self.y})'

a = Point(1, 2)

print(str(a))
print(repr(a))

```
