````python
class Flyable:
    def fly(self):
        print("날아갑니다")

class Swimmable:
    def swim(self):
        print("수영합니다.")

class Duck(Flyable, Swimmable):

    def do_everything(self):
        self.fly()
        self.swim()

duck = Duck()

duck.do_everything()
duck.fly()
duck.swim()
````

````python
class Animal:
    def __init__(self, name):
        self.name = name
        print(f'Animal.__init__ 호출 : {self.name}')

    def introduce(self):
        return f'저는 {self.name}입니다.'

class Dog(Animal):

    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed
        print(f'Dog.__init__ 호출 : {self.name}')

    def introduce(self):
        base = super().introduce()
        return f'{base} 그리고 저는 {self.breed}입니다.'

class Huskey(Dog):

    def __init__(self, name, breed, color):
        super().__init__(name, breed)
        self.color = color
        print(f'Huskey.__init__ 호출 : {self.name}')

    def introduce(self):
        base = super().introduce()
        return f'{base} 그리고 털 색은 {self.color}입니다.'


h = Huskey('아롱이', '진돗개', '흰색')

print(h.introduce())
````

# MRO(Method Resolution Order)



````python
class A:
    def hello(self):
        print("A")

class B(A):
    def hello(self):
        print("B")

class C(A):
    def hello(self):
        print("C")

class D(B, C):
    pass

d = D()
d.hello()

print(D.__mro__)
````

단일 상속
- 자식 클래스 -> 부모 클래스 -> 조상 클래스 -> object

다중 상속
- 자식이 부모보다 앞에 온다: 자식클래스는 항상 부모 클래스보다 먼저 탐색됨
- 부모 클래스간의 순서가 유지된다: classC(A, B)에서는 A가 항상 B보다 먼저 탐색됨


# 추상 클래스

````python
class Shape:

    def area(self):
        raise NotImplementedError('area()를 구현하지 않았어? 자신있어?')
````

````python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# shape = Shape()

class Circle(Shape):

    def __init__(self, radius):
        super().__init__()
        self.radius = radius

    def area(self):
        import math
        return math.pi * (self.radius ** 2)

print(Circle(5).area())
````

# Mixin

Mixin 독립적인 주체가 되기보다는 다른 클래스에 특정 기능을 추가하기 위해 설계된 작은 클래스를 의미합니다.

클래스의 이름 끝에다가 Mixin, Mixin 다중 상속을 활용합니다.

`부모클래스 - 자식클래스 -> IS-A `

Mixin 새로운 능력을 부여한다.

````python
class SoundMixin:
    """소리내는 기능"""
    def bark(self):
        print(f'{self.name}: 멍멍')

class RunMixin:
    """달리는 기능"""
    def run(self):
        print(f'{self.name}이 빠르게 달립니다. 헥헥')

class ValidMixin:
    """이름 유효성 검사기능"""
    def is_valid(self):
        return len(self.name) > 0


class Dog(SoundMixin, RunMixin, ValidMixin):
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

dog = Dog("까망이", "진돗개")

dog.bark()
dog.run()
````
