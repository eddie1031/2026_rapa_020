```python
class Car:

    def __init__(self, name):
        self.name = name

    def drive(self):
        print(f"{self.name}이가 붕붕")

car = Car('붕붕')
car.drive()
```

```python
class Student:
    school_name: str = "RAPA 백엔드반"
    attend_count: int = 0

    def __init__(self, name):
        self.name = name
        Student.attend_count += 1

print(Student.school_name)

ms = Student('김민수')
print(ms.attend_count)

yh = Student('박영희')
print(yh.attend_count) # 2
print(ms.attend_count) # 2


```
