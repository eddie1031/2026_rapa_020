# 자료형(Data Type)

자료형은 기능적 관점, 가변성 관점에 따라 크게 2가지로 구분해볼 수 있다.


## 기능적 분류
- 숫자 자료형(Numeric Type)
- 문자 자료형(Text Type)
- 시퀀스 자료형(Sequence Type) -> 변경 가능한 시퀀스(가변성), 변경 불가능한 시퀀스(불변성)
- 매핑 자료형(Mapping Type) -> 키-값
- 집합 자료형(Set Type) -> 중복 없는 가변 집합
- 논리 자료형(Boolean Type)
- 바이트 자료형(Bytes Type) ->
- None 자료형(None Type) ->

## 논리 연산자

`&&`, `||` 를 사용하지 않습니다.

`and`, `or`, `not`


```python
x = 7

if 1 < x < 10:
    pass
```

```python
x = 0
result = x != 0 and ( 10 / x > 1 )

print(result)
```

```python
def check_true():
    print("check true 호출!")
    return True

def check_false():
    print("check false 호출!")
    return False

result1 = check_false() and check_true()
print(result1)
```

## 식별 연산자
`is` `id()`

```python
a = [1,2,3]
b = [1,2,3]
c = a

print(a == b)
print(a is b)
print(a is c)
```

## 바다코끼리연산자  :=





```python
n = len('파이썬')
if n > 2:
    print(f'길이 {n}은 2보다 크다')

if ( n := len("파이썬") ) > 2:
    print(f'길이 {n}은 2보다 크다')
```
