# 모듈(Module)

`import` 다른 모듈의 내용을 현재 코드에서 사용할 수 있도록 불러오는 구문입니다.

Python 인터프리터는 `import`를 만나면 다음의 순서대로 관련 모듈을 탐색합니다.

1. 실행중인 스크립트의 디렉토리를 먼저 탐색
2. PYTHONPATH 환경변수에 지정된 경로 탐색
3. 표준 라이브러리 경로 탐색

모듈을 한 번 불러왔으면 이후에는 캐싱(sys.modules)이됩니다. 다른곳에서 import해도 재실행되지 않습니다.

```python
import math
import os
import sys
```

## from ... import

`from ... import` 모듈에서 특정이름(클래스, 함수, 변수 ...)만 선택적으로 불러오는 방식입니다.


```python
from math import sqrt, pi, floor

print(sqrt(25))
print(pi)
```
