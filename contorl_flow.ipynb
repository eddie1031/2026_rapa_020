# Iterable(반복 가능한 객체)

반복 가능한 객체란, for 루프 등에서 원소를 하나씩 꺼낼수 있는 객체를 의미합니다.
구체적으로는 `__iter__()` 메서드를 구현한 모든 객체가 이에 해당할 수 있습니다.
리스트, 튜플, 문자열, 딕셔너리, 집합, range, 파일객체, 제너레이터 같은 객체들이 Iterable한 객체입니다.

Iterable이랑 Iterator는 서로 다른 것입니다. Iterable은 반복 가능한 객체, Iterator는 Iterable에서 iter() 함수로 생성되어서 next() 함수로 원소를 하나씩 반환하는 반복자 패턴의 구현 객체입니다.



```python
my_list = [10, 20, 30]
my_iter = iter(my_list)
print(type(my_iter))

while True:
    try:
        element = next(my_iter)
        print(element)
    except StopIteration:
        break

class EvenNumber:

    def __init__(self, max_val):
        self.max_val = max_val
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        self.current += 2
        if self.current > self.max_val:
            raise StopIteration
        return self.current

for n in EvenNumber(10):
    print(n, end=' ')

print()
```

# enumerate()

`enumerate()`는 Iterable을 순회할때 원소와 함꼐 인덱스(순번)를 동시에 제공하는 내장 함수.

```python
fruits = [ '사과', '배', '포도', '딸기' ]

for i, fruit in enumerate(fruits):
    print(f'index: {i}, fruit: {fruit}')

for fruit in enumerate(fruits):
    print(f'result: {fruit}')
```

# zip()

`zip()`은 두 개 이상의 Iterable을 같은 인덱스의 원소끼리 묶어 (원소1, 원소2, 원소3, ...) 형태의 튜플을 반환하는 내장함수


```python
names = [ 'maki', 'reina', 'matsumoto' ]
scores = [ 1, 20, 40 ]

for name, score in zip(names, scores):
    print(f'name: {name}, score: {score}')


a = [1,2,3,4,5]
b = ['a', 'b', 'c']

print(list(zip(a,b)))


persons = dict(zip(names, scores))
print(persons)
```
