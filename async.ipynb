# 비동기 프로그래밍

비동기 프로그래밍은 프로그램이 시간이 오래걸리는 작업을 기다리는 동안 다른 작업을 함께 진행할 수 있도록 만드는 프로그래밍 방식.

Python에서는 비동기 프로그래밍 모듈, `asyncio` -> 이벤트 루프, 코루틴, 태스크, 퓨처



# 동시성과 병렬성

- 동시성(Concurrency): 여러 작업이 동시에 진행되는 것처럼 보이도록 구성된 실행 모델, 실제로 동시에 실행되는가? x -> 동시에 처리되는것처럼 관리되는가? -> 여러 작업이 공유 작업에 접근할때 경쟁 상태(Race Condition), 잠금(lock), 세마포어, 뮤텍스, 데드락, 기아상태(Starvation)

- 병렬성(Parallelism): 여러 작업이 물리적으로 동시에 실행되는것을 의미합니다. 다중코어 CPU에서 여러개의 프로세서를 사용할때 사용가능한 방식.



## asyncio

`asyncio`는 Python 표준 라이브러리에서 제공하는 비동기 프로그래밍 프레임워크


asyncio는 코루틴(Coroutine)을 실행하고 이벤트 루프(Event Loop)를 통해서 여러 작업을 스케줄링합니다.






## 구조

### 코루틴

`async def` 정의된 비동기 함수를 의미.


## Event Loop

asyncio의 실행 엔진입니다.

동기 프로그래밍 -> 함수가 호출 -> 호출 스택 -> 위에서 아래로 실행 -> 시간이 오래걸리는 작업을 만나면 -> 기다림 -> 명령을 끝까지 수행

asyncio에서는 코루틴 await -> 이벤트 루프에게 제어권 넘김 -> 현재 작업이 다시 실행 가능한 상태가 될때까지 다른 작업을 수행함
