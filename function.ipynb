# 매개변수에 언패킹

````python
def add(a,b,c):
    return a + b + c

val_1 = [1,2,3]
val_2 = 4,5,6

print(add(val_1[0], val_1[1], val_1[2]))
print(add(*val_1))

print(add(*val_2))
````

````python
def describe(name, age, city):
    print(f"""
    이름: {name}
    나이: {age}
    도시: {city}
    """)

info = { 'name': "PengXin", 'age': 21, 'city': "Xi'an" }

describe(**info)
````

````python
def func(a,b,c,d):
    return a + b + c + d

positional = [ 1, 2, ]
keyword = { 'c': 3, 'd': 4 }

print(func(*positional, **keyword))
````

````python
def log_wrapper(func, *args, **kwargs):
    print(f'전처리')
    result = func(*args, **kwargs)
    print(f'후처리')
    return result

log_wrapper(describe, name='PengXin', age=21, city="Xi'an")
result = log_wrapper(add, 1,2,3)
print(result)
````
